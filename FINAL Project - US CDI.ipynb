    - Total: An estimated 30.3 million people have diabetes (9.4 percent of the U.S. population)
    - Diagnosed: An estimated 23.1 million people have been diagnosed with diabetes (7.2 percent of the U.S. population)
    - Undiagnosed: An estimated 7.2 million adults, ages 18 years or older are undiagnosed (23.8 percent of people with diabetes)

There is an obvious difference in alcohol prevalence for males and females so we will predict using other chronic disease factors.

```python
X = merged_df_gender[[
 'DataValue_diabetes',

 'DataValue_tobacco',
 'DataValue_obesity',
 'DataValue_physical_activity',
 'DataValue_cardio']].values

y = merged_df_gender['Stratification1'].values
```

```python
#Choose features based on chi squared test. 
features = [
 'DataValue_diabetes',

 'DataValue_tobacco',
 'DataValue_obesity',
 'DataValue_physical_activity',
 'DataValue_cardio']
print(features)
selector = SelectPercentile(chi2)
X_new = selector.fit_transform(X, y)

print("P Values:", selector.pvalues_)
```

```python
XTrain, XTest, yTrain, yTest = train_test_split(X, y, random_state=1, test_size=0.3)
```

```python
knearest_model = KNeighborsClassifier(n_neighbors = 10)
knearest_model.fit(XTrain, yTrain)

y_pred = knearest_model.predict(XTest)

print('Confusion Matrix:')
print(metrics.confusion_matrix(y_true = yTest, y_pred = y_pred))

print('Accuracy = ', metrics.accuracy_score(y_true = yTest, y_pred = y_pred))
```

```python
from sklearn import svm
svm_model = svm.SVC(kernel='rbf',C=1,gamma='scale')
svm_model.fit(XTrain, yTrain)

y_pred = svm_model.predict(XTest)

print('Confusion Matrix:')
print(metrics.confusion_matrix(y_true = yTest, y_pred = y_pred))

print('Accuracy = ', metrics.accuracy_score(y_true = yTest, y_pred = y_pred))
```

Support Vector Machine (SVM ) model has a better accuracy than the k-neighbors classifier.
